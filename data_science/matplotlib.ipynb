```python
x = np.linspace(0, 50, 100)
y1 = x * np.linspace(100, 200, 100)
y2 = x * np.random.randint(0, 200, 100)
plt.plot(x, y1, c='r', linestyle='--', linewidth=2, label="x-y1")
# style of combining color, linestyle together
plt.plot(x, y2, 'g-', label="x-y2")
plt.title("Basic Plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xlim(0, 60)
plt.ylim(0,12000)
plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```

### subplots

#### 1) plt.subplot()

```python
plt.figure(figsize=(12,6), dpi=80)
# scatter chat
plt.subplot(2,3,1)
plt.title("scatter chat")
x = np.linspace(0, 5, 100)
y = x + np.random.randn(len(x))
plt.xlabel("Time/s")
plt.ylabel('Value')
plt.scatter(x, y)

# bar char
plt.subplot(2,3,2)
plt.title("bar chat")
plt.xlabel("Categories")
plt.ylabel('Number')
categories = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
grades = np.random.randint(20, 60, len(categories))
plt.bar(categories, grades, align='center', width=0.5, alpha=0.8)

# Histogram char
plt.subplot(2,3,3)
x = np.random.randint(1,1000, 100)
plt.title("Histogram chat")
plt.xlabel("Values")
plt.ylabel('Counts')
plt.hist(y, bins=30, color='g')

# pie char
plt.subplot(2,3,4)
# plt.tight_layout(2)
data = sorted(grades)
explode = [0.1 if x > 50 else 0 for x in data]
plt.pie(data, labels=categories, autopct='%1.1f%%', explode=explode, radius=1.2)

# boxplot
plt.subplot(2,3,5)
plt.tight_layout(2)
plt.title("box plot")
plt.boxplot(np.random.randn(10))
plt.show()
```

#### 2) add_subplot()

```python
fig = plt.figure(figsize=(10,6))
for i in range(1, 5):
    ax = fig.add_subplot(2,2, i)
    plt.tight_layout()
    x = np.random.randint(-2, 2, 10)
    y = np.sin(x)
    ax.set(title=f"y-{i}", xlabel='x', ylabel="sin(x)")
    ax.plot(x, y)
plt.show()
```

### plt.add_axes()

```python
fig = plt.figure(figsize=(10,8))
x = np.linspace(-0.5, 1., 1000)
plt.plot(x, x**2)
plt.plot(x, np.sin(x))
plt.xlim=(-0.5, 1.0)
plt.ylim=(-0.5, 1.2)
plt.title("Plot inside plot", fontsize=20)
plt.tight_layout()

inner_ax = fig.add_axes([0.2, 0.55, 0.35, 0.35])
inner_ax.plot(x, x**2, label='line1')
inner_ax.plot(x, np.sin(x), label='line2')
inner_ax.legend(loc='lower right')
inner_ax.set(title="Zoom-in", xlim=(-.2, .2), ylim=(-.01,.02 ))
plt.show()
```
